```python
# Crear copia para limpieza
df_co597_copy = df_co597.copy()


# Limpiar las columnas num√©ricas
columnas_numericas_limpiar = ['valor', 'Textbox6','Textbox16']

for col in columnas_numericas_limpiar:
    if col in df_co597_copy.columns:
        df_co597_copy[col] = df_co597_copy[col].apply(sl.limpiar_valor_numerico)
```

```python
# Renombrar las columnas principales para claridad
df_co597_copy = df_co597_copy.rename(columns={
    'nombreZonaAreaGeoMundo': 'area',
    'Pais': 'pais',
    'tipo_desplazamiento': 'desplazamiento',
    'year': 'a√±o',
    'Unidad': 'unidad',
    'Valor': 'valor',
    'Textbox4':'Textbox4',
    'Textbox6':'Textbox6',
    'Textbox16':'Textbox16'
})

# Seleccionar las columnas relevantes
columnas_relevantes = ['area','pais','desplazamiento','a√±o','unidad','valor']

df_co597_copy = df_co597_copy[columnas_relevantes].copy()
df_co597_copy.head()
```

```python
# Separar por tipo de unidad
df_co597_toneladas = df_co597_copy[
    df_co597_copy['unidad'] == 'Miles de toneladas'
].copy()

df_co597_ton_km = df_co597_copy[
    df_co597_copy['unidad'] == 'Millones de toneladas-kil√≥metro'
].copy()

df_co597_toneladas.columns = ['area','pais','desplazamiento','a√±o','unidad','toneladas(miles)']
df_co597_ton_km.columns = ['area','pais','desplazamiento','a√±o','unidad','ton-km(millones)']

# Eliminar la columna unidad ya que ahora es redundante
df_co597_toneladas = df_co597_toneladas.drop('unidad', axis=1)
df_co597_ton_km = df_co597_ton_km.drop('unidad', axis=1)
```

```python
df_co597_final = df_co597_ton_km.merge(
    df_co597_toneladas,
    on=['area','pais','desplazamiento','a√±o'],
    how='inner',
    validate='one_to_one'  # ver si hay duplicados en las claves
)

# Eliminar NaN correspondientes a totales
if 'pais' in df_co597_final.columns:
        before = len(df_co597_final)
        df_co597_final = df_co597_final.dropna(subset=['pais'])
        print(f"Eliminadas {before - len(df_co597_final)} filas con NaN en columna 'pais'")

df_co597_final.head(10)
```

```python
# Guardar el archivo limpio
archivo_co597_clean = os.path.join(ruta_processed, 'CO597_transporte_mercancias_internacional.csv')
df_co597_final.to_csv(archivo_co597_clean, index=False, encoding='utf-8')

print(f"‚úÖ Archivo guardado: {archivo_co597_clean}")
```
