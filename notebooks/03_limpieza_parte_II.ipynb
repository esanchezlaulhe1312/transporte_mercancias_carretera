```python
# Crear copia para limpieza
df_co516_copy = df_co516.copy()

# Identificar qu√© representa cada columna Textbox mirando los patrones
print("üîç Identificando contenido de columnas Textbox...")
print(f"Textbox10 parece ser: superficie en m¬≤ (valores como 14.569)")
print(f"Textbox27 parece ser: descripci√≥n del subtotal")
print(f"Textbox28 parece ser: total de superficie")

# Limpiar las columnas num√©ricas
columnas_numericas_limpiar = ['Textbox10', 'Textbox28', 'Textbox33', 
                              'Textbox22', 'Textbox24', 'Textbox29', 
                              'Textbox34', 'Textbox14', 'Textbox19', 
                              'Textbox20', 'Textbox21']

for col in columnas_numericas_limpiar:
    if col in df_co516_copy.columns:
        df_co516_copy[col] = df_co516_copy[col].apply(sl.limpiar_valor_numerico)
```

```python
# 3. Renombrar las columnas principales para claridad
df_co516_copy = df_co516_copy.rename(columns={
    'year': 'a√±o',
    'nombreAliMkt_tipoInst': 'tipo_instalacion',
    'nombreAliMkt_Inst': 'subtipo_instalacion',
    'Textbox10': 'superficie_m2'
})

# 4. Seleccionar las columnas relevantes
columnas_relevantes = ['a√±o', 'comunidad', 'provincia', 
                      'tipo_instalacion', 'subtipo_instalacion', 
                      'superficie_m2']

df_co516_copy = df_co516_copy[columnas_relevantes].copy()
df_co516_copy.head()
```

```python
# 5. Eliminar filas con valores nulos en superficie_m2
print(f"\n‚ùå Eliminando {df_co516_copy['superficie_m2'].isnull().sum()} filas sin superficie...")
df_co516_copy = df_co516_copy.dropna(subset=['superficie_m2'])

# 6. Verificar el resultado
print(f"\n‚úÖ Dataset limpio:")
print(f"- Dimensiones: {df_co516_copy.shape}")
print(f"- Superficie total: {df_co516_copy['superficie_m2'].sum():,.0f} m¬≤")

# Ver resumen por tipo de instalaci√≥n
print(f"\nüìä Superficie por tipo de instalaci√≥n:")
resumen_tipo = df_co516_copy.groupby('tipo_instalacion')['superficie_m2'].agg(['sum', 'count'])
resumen_tipo['sum'] = resumen_tipo['sum'].round(0)
print(resumen_tipo)
```

```python
# Guardar el archivo limpio
archivo_co516_clean = os.path.join(ruta_processed, 'CO516_superficies_logisticas_clean.csv')
df_co516_copy.to_csv(archivo_co516_clean, index=False, encoding='utf-8')

print(f"‚úÖ Archivo guardado: {archivo_co516_clean}")

# Crear un resumen m√°s detallado por comunidad aut√≥noma
print("\nüìä TOP 5 Comunidades por superficie log√≠stica total:")
superficie_por_ccaa = df_co516_copy.groupby('comunidad')['superficie_m2'].sum().sort_values(ascending=False)
for ccaa, superficie in superficie_por_ccaa.head().items():
    print(f"  {ccaa}: {superficie:,.0f} m¬≤ ({superficie/1000000:.1f} millones m¬≤)")

# Resumen del proceso de limpieza
print("\nüìä RESUMEN DE LIMPIEZA - CO516:")
print(f"- Registros originales: 10,402")
print(f"- Registros con superficie nula: 6,586 (63.3%)")
print(f"- Registros limpios finales: {len(df_co516_copy)}")
print(f"- Tipos de instalaci√≥n: 5")
print(f"- Subtipos de instalaci√≥n: 26")
print(f"- Provincias con datos: {df_co516_copy['provincia'].nunique()}")
print(f"- Superficie total: {df_co516_copy['superficie_m2'].sum()/1000000:.1f} millones m¬≤")
print(f"- A√±o de los datos: 2017 (diciembre)")
```

```python
# Crear copia para limpieza
df_co597_copy = df_co597.copy()


# Limpiar las columnas num√©ricas
columnas_numericas_limpiar = ['valor', 'Textbox6','Textbox16']

for col in columnas_numericas_limpiar:
    if col in df_co597_copy.columns:
        df_co597_copy[col] = df_co597_copy[col].apply(sl.limpiar_valor_numerico)
```

```python
# Renombrar las columnas principales para claridad
df_co597_copy = df_co597_copy.rename(columns={
    'nombreZonaAreaGeoMundo': 'area',
    'Pais': 'pais',
    'tipo_desplazamiento': 'desplazamiento',
    'year': 'a√±o',
    'Unidad': 'unidad',
    'Valor': 'valor',
    'Textbox4':'Textbox4',
    'Textbox6':'Textbox6',
    'Textbox16':'Textbox16'
})

# Seleccionar las columnas relevantes
columnas_relevantes = ['area','pais','desplazamiento','a√±o','unidad','valor']

df_co597_copy = df_co597_copy[columnas_relevantes].copy()
df_co597_copy.head()
```

```python
# Separar por tipo de unidad
df_co597_toneladas = df_co597_copy[
    df_co597_copy['unidad'] == 'Miles de toneladas'
].copy()

df_co597_ton_km = df_co597_copy[
    df_co597_copy['unidad'] == 'Millones de toneladas-kil√≥metro'
].copy()

df_co597_toneladas.columns = ['area','pais','desplazamiento','a√±o','unidad','toneladas(miles)']
df_co597_ton_km.columns = ['area','pais','desplazamiento','a√±o','unidad','ton-km(millones)']

# Eliminar la columna unidad ya que ahora es redundante
df_co597_toneladas = df_co597_toneladas.drop('unidad', axis=1)
df_co597_ton_km = df_co597_ton_km.drop('unidad', axis=1)
```

```python
df_co597_final = df_co597_ton_km.merge(
    df_co597_toneladas,
    on=['area','pais','desplazamiento','a√±o'],
    how='inner',
    validate='one_to_one'  # ver si hay duplicados en las claves
)

# Eliminar NaN correspondientes a totales
if 'pais' in df_co597_final.columns:
        before = len(df_co597_final)
        df_co597_final = df_co597_final.dropna(subset=['pais'])
        print(f"Eliminadas {before - len(df_co597_final)} filas con NaN en columna 'pais'")

df_co597_final.head(10)
```

```python
# Guardar el archivo limpio
archivo_co597_clean = os.path.join(ruta_processed, 'CO597_transporte_mercancias_internacional.csv')
df_co597_final.to_csv(archivo_co597_clean, index=False, encoding='utf-8')

print(f"‚úÖ Archivo guardado: {archivo_co597_clean}")
```
